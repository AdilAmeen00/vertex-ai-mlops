## Structured Approach

```python
vertex_arch = dict(name = 'Vertex AI', l1 = dict(
    shape = (4, 1),
    parts = [
        dict(name = 'Console UI', l2 = dict()),
        dict(name = 'Workbench', l2 = dict(
            shape = (1, 3),
            parts = [
                dict(name = 'Instances', l3 = dict()),
                dict(name = 'User-Managed Notebooks', l3 = dict()),
                dict(name = 'Managed Notebooks', l3 = dict())
            ]
        )),
        dict(name = 'Model Development', l2 = dict(
            shape = (1, 3),
            parts = [
                dict(name = 'Data', l3 = dict(
                    shape = (3, 1),
                    parts = [
                        dict(name = 'Datasets'),
                        dict(name = 'Feature Store'),
                        dict(name = 'Labeling Tasks')
                    ]
                )),
                dict(name = 'Training', l3 = dict(
                    shape = (4, 1),
                    parts = [
                        dict(name = 'Custom Jobs'),
                        dict(name = 'Training Pipelines'),
                        dict(name = 'Hyperparameter Tuning Jobs'),
                        dict(name = 'NAS Jobs')
                    ]
                )),
                dict(name = 'Serving', l3 = dict(
                    shape = (3, 1),
                    parts = [
                        dict(name = 'Online Prediction'),
                        dict(name = 'Batch Predictions'),
                        dict(name = 'Matching Engine')
                    ]
                ))
            ]
        )),
        dict(name = 'Pipelines', l2 = dict(
            shape = (1, 4),
            parts = [
                dict(name = 'Template Gallery', l3 = dict()),
                dict(name = 'Your Templates', l3 = dict()),
                dict(name = 'Schedules', l3 = dict()),
                dict(name = 'Runs', l3 = dict())
            ]
        ))
    ]
))
```

```python
graph_attr = dict(bgcolor = 'white', labelloc = 'topCenter', fontcolor = 'black')
node_attr = dict(shape = 'rectangle', style = 'filled,rounded', fontcolor = gcolors['blue'][0], fillcolor = gcolors['blue'][3])
subgraph_attr = dict(style = 'filled,rounded', labelloc = 'topCenter', fontcolor = gcolors['blue'][0], fillcolor = gcolors['blue'][8])
sub_subgraph_attr = dict(style = 'filled,rounded', labelloc = 'topCenter', fontcolor = gcolors['blue'][0], fillcolor = gcolors['blue'][4])
```

```python
# settings
graph_levels = ['l1', 'l2', 'l3']

# BEGIN GRAPH
g = graphviz.Digraph()
g.attr(**graph_attr)

# iterate level 1: l1
if vertex_arch['l1']:
    g.attr(label = vertex_arch['name'])
    nodes = []
    for n1, l1 in enumerate(vertex_arch['l1']['parts']):

        # add nodes
        if not l1['l2'] or 'l2' not in graph_levels:
            g.node(name = f'node_l1_{n1}', label = l1['name'], **node_attr)
            nodes.append(f'node_l1_{n1}')

        # or add subgraph: iterate level 2: l2
        elif l1['l2'] and 'l2' in graph_levels:
            with g.subgraph(name = f'cluster_l1_{n1}') as s:
                s.attr(label = l1['name'], **subgraph_attr)
                for n2, l2 in enumerate(l1['l2']['parts']):

                    # add nodes
                    if not l2['l3'] or 'l3' not in graph_levels:
                        s.node(name = f'node_l1_{n1}_l2_{n2}', label = l2['name'], **node_attr)
                        nodes.append(f'node_l1_{n1}_l2_{n2}')


                    # or add subgraph: iterate level 3: l3
                    elif l2['l3'] and 'l3' in graph_levels:
                        with s.subgraph(name = f'cluster_l2_{n2}') as ss:
                            ss.attr(label = l2['name'], **sub_subgraph_attr)
                            for n3, l3 in enumerate(l2['l3']['parts']):

                                # add nodes - no deeper levels to consider
                                ss.node(name = f'node_l1_{n1}_l2_{n2}_l3_{n3}', label = l3['name'], **node_attr)
                                nodes.append(f'node_l1_{n1}_l2_{n2}_l3_{n3}')

    # enforce order with invisible edges between nodes
    for n, node in enumerate(nodes):
        if n > 0: g.edge(nodes[n-1], node, style = 'invis')
                            
# END GRAPH
g
```
